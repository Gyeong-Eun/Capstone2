```python
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium import webdriver


driver = webdriver.Chrome(service=Service(chromedriver_autoinstaller.install()), options=options)

# 기사 url리스트 / 기사 제목 리스트 / 언론사 리스트 / 키워드 리스트
all_hrefs = []
all_titles = []
all_companies = []
all_keywords = []

# 이미 수집한 (URL, 제목) 쌍을 추적하기 위해 리스트 사용
collected_pairs = list(zip(all_hrefs, all_titles, all_companies))  # 이미 수집한 (URL, 제목, 언론사) 쌍

# 변화가 없을 때를 추적하는 변수
no_change_count = 0
max_no_change = 15  # 최대 반복 횟수

# 키워드별 URL을 돌면서 제목, 기사 URL 수집
for keyword in keyword_list:
    driver.get(company_url)
    time.sleep(1)

    # 검색창 요소 찾기
    search_box = driver.find_element(By.NAME, 'query')

    # 현재 입력된 내용을 지우기
    search_box.clear()

    # 검색어 입력
    search_term = keyword  # 검색할 단어
    search_box.send_keys(search_term)

    # 엔터 키를 눌러 검색 실행
    search_box.send_keys(Keys.RETURN)

    # 페이지 로딩 대기 (최대 10초 동안 a.news_tit 요소가 나타날 때까지 대기)
    time.sleep(3)

    # 반복 멈출 길이 조건
    count = 0

    while True:  # 페이지 끝까지 스크롤하기 위한 무한 루프
        # li.bx 요소들을 모두 가져옴
        li_elements = driver.find_elements(By.CSS_SELECTOR, "div.news_area")

        # 새롭게 발견된 (URL, 제목, 언론사, 키워드) 쌍 추출
        new_pairs = [
            (element.find_element(By.CSS_SELECTOR, "a.news_tit").get_attribute('href'),
             element.find_element(By.CSS_SELECTOR, "a.news_tit").text,
             '세계일보',
             keyword)
            for element in li_elements
            if keyword in element.find_element(By.CSS_SELECTOR, "a.news_tit").text  # 키워드가 제목에 포함될 때
        ]

        # 중복되지 않는 (URL, 제목, 언론사, 키워드) 쌍만 필터링
        filtered_pairs = [pair for pair in new_pairs if pair not in collected_pairs]

        # 키워드가 제목에 포함된 경우 count 증가
        count += len(filtered_pairs)

        # 필터링된 (URL, 제목, 언론사, 키워드) 쌍을 기존 리스트에 추가
        all_hrefs.extend([pair[0] for pair in filtered_pairs])
        all_titles.extend([pair[1] for pair in filtered_pairs])
        all_companies.extend([pair[2] for pair in filtered_pairs])
        all_keywords.extend([pair[3] for pair in filtered_pairs])

        print("언론사 이름: ", '세계일보')
        print("키워드 명: ", keyword)
        print("총 기사 수: ", count)

        # 중복 방지를 위해 새롭게 수집된 쌍을 추가
        collected_pairs.extend(filtered_pairs)

        # 페이지 높이를 확인하고 스크롤
        last_height = driver.execute_script("return document.body.scrollHeight")
        driver.execute_script("window.scrollBy(0, 2000);")  # 스크롤
        time.sleep(1)  # 페이지 로딩 대기
        new_height = driver.execute_script("return document.body.scrollHeight")

        # 기사가 15개 이상이면 종료
        if count >= 15:
            break

        # 길이가 변화하지 않으면 no_change_count 증가
        if len(filtered_pairs) == 0:  # 새로 추가된 게 없으면
            no_change_count += 1
        else:
            no_change_count = 0  # 변화가 있으면 초기화

        # 변화가 15번 동안 없으면 종료
        if no_change_count >= max_no_change:
            print('변화가 없습니다.')
            break

        # 더이상 스크롤 불가 종료
        if last_height == new_height:
            print('더 이상 스크롤할 수 없습니다.')
            break

        # 새 높이로 업데이트
        last_height = new_height

print(len(all_hrefs))
```

```python
import pandas as pd
df = pd.read_excel('/content/세계일보_2023키워드_중복제거.xlsx')
# '언론사', '기사_url', '제목', '키워드' 컬럼의 데이터를 각각 리스트로 가져오기
all_companies = df['언론사'].tolist()
all_hrefs = df['기사_url'].tolist()
all_titles = df['제목'].tolist()
all_keywords = df['키워드'].tolist()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, NoSuchElementException, UnexpectedAlertPresentException
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 저장된 데이터
new_hrefs = []
new_titles = []
all_content = []
new_company = []
new_keyword = []

# 세계일보 본문과 제목 선택자
selectors = {
    'content_selector': '#wps_layout1_box2',
    'title_selector': '#title_sns'  # 제목의 CSS 선택자 (예시로 h1.title 사용)
}

driver = webdriver.Chrome(service=Service(chromedriver_autoinstaller.install()), options=options)
driver.set_page_load_timeout(60)  # 페이지 로드 타임아웃 60초 설정

max_retries = 2  # 최대 재시도 횟수 설정

for i in range(len(all_hrefs)):
    retries = 0  # 재시도 횟수 초기화
    while retries < max_retries:
        try:
            driver.get(all_hrefs[i])  # 페이지 로드
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'body')))

            # 페이지가 로드된 후 스크롤하여 추가 콘텐츠 로드
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # 잠시 대기

            # 본문과 제목이 나타날 때까지 기다림
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selectors['content_selector'])))
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selectors['title_selector'])))
            break  # 페이지 로드가 성공했으면 while 루프를 탈출

        except TimeoutException:
            print(f"페이지 로딩 실패 (타임아웃): {all_hrefs[i]}")
            retries += 1  # 재시도 횟수 증가
            time.sleep(3)  # 잠시 대기 후 재시도
            if retries == max_retries:
                print(f"최대 재시도 초과: {all_hrefs[i]}")
                break  # 최대 재시도 시 while 루프를 탈출하여 다음 기사로 넘어감

        except Exception as e:
            print(f"예상치 못한 오류 발생: {i+1}번째 기사에서 에러 발생 - {str(e)}")
            break  # 다른 예외가 발생해도 while 루프를 탈출하고 다음 기사로 넘어감

    # 본문 및 제목 추출
    try:
        # 본문 추출
        content = driver.find_element(By.CSS_SELECTOR, selectors['content_selector']).text
        all_content.append(content)

        # 제목 추출
        title = driver.find_element(By.CSS_SELECTOR, selectors['title_selector']).text
        new_titles.append(title)

        # URL, 언론사, 키워드 추가
        new_hrefs.append(all_hrefs[i])
        new_company.append('세계일보')
        new_keyword.append(all_keywords[i])

        # 데이터 저장
        save_data()  # 데이터 저장 함수 호출

    except NoSuchElementException as e:
        print(f"요소를 찾을 수 없습니다: {i+1}번째 기사 - {str(e)}")
        continue  # 해당 기사를 건너뛰고 다음 기사로 넘어감

    except TimeoutException:
        print(f"페이지 로딩 실패 (타임아웃): {all_hrefs[i]}")
        continue  # 타임아웃 발생 시 다음 기사로 넘어감

    except UnexpectedAlertPresentException:
        alert = Alert(driver)
        print(f"경고창 발견: {alert.text} - {i+1}번째 기사")
        alert.accept()  # 경고창 닫기
        continue  # 경고창 발생 시 다음 기사로 넘어감

    except Exception as e:
        print(f"예상치 못한 오류 발생: {i+1}번째 기사에서 에러 발생 - {str(e)}")
        continue  # 예외가 발생하면 다음 기사로 넘어감

# 최종적으로 수집된 데이터 저장
save_data()
driver.quit()

# 최종적으로 수집한 본문 및 메타 데이터 길이 출력
print(f"총 수집된 url 수: {len(new_hrefs)}")
print(f"총 수집된 제목 수: {len(new_titles)}")
print(f"총 수집된 본문 수: {len(all_content)}")
print(f"총 수집된 언론사 수: {len(new_company)}")
print(f"총 수집된 키워드 수: {len(new_keyword)}")
```
